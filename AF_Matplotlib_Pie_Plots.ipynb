```python
import ast
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df.head()
```

```python
df['job_work_from_home'].value_counts()
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie')

# pie chart needs a lot of cleanup

plt.ylabel('')
plt.title('Work from Home Status')
plt.show()
```

```python
# same chart for job_title_short:

df['job_title_short'].value_counts().plot(kind='pie')
plt.ylabel('')
plt.title('Job Title Split')
plt.show()

# it's hard to compare the distances between these titles
```

```python
# same data is much clearer with bar

df['job_title_short'].value_counts().plot(kind='barh')
plt.ylabel('')
plt.title('Job Title Split')
plt.gca().invert_yaxis()
plt.show()
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie', startangle=90, autopct='%1.2f%%')
plt.ylabel('')
plt.title('Job Title Split')
plt.show()

# startangle = 90 aligns the True slice to the center right axis
# autopct='%1f%%' is to display % in data labels / % means 'begins with' / 1 is min width of number preceding dec.point / .1 means 1 place after dec.point / f signifies float dtype 
```

```python
fig, ax = plt.subplots(1,3)

dict_column = {
    'job_work_from_home' : 'Work from Home',
    'job_no_degree_mention': 'Job Degree Req',
    'job_health_insurance' : 'Health Insurance offered'
}
for i, (column, title) in enumerate(dict_column.items()):
    ax[i].pie(df[column].value_counts(), startangle=90, autopct='%1.1f%%', labels=['False', 'True'])
    ax[i].set_title(title)

plt.show()
```

## Data Cleaning 
# iloc() --> loc()

```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

```

```python
df.iloc[0]

# iloc locks the index and returns the mentioned values
```

```python
df.iloc[0:10, 0:3]

# this returns the first 10 rows and first 3 columns
```

```python
# .loc() on the other hand is a more efficient function, especially with the columns parameter (slicing of columns is doable); additionally we remove the null values

df.loc[:, 'salary_rate' : 'salary_year_avg'].dropna(subset='salary_year_avg')
```

```python
# Get the first 5 columns of the DataFrame and all the rows.

df.iloc[:, 0:5]
```

## Plotting monthly job postings to determine job demand
### Task:- Plot the following graph:


```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_usa = df[df['job_country'] == 'United States'].copy()

df_usa
```

```python
df_usa['job_posted_month'] = df_usa['job_posted_date'].dt.strftime('%B') 

# strftime creates a string time format that helps us to categorize the month from the date; '%B' is the specified format (from API Reference)

df_usa
```

```python
# create a pivot table indexing the job_posted_month

df_usa_pivot = df_usa.pivot_table(
    values= 'salary_year_avg',
    index= 'job_posted_month',
    columns= 'job_title_short',
    aggfunc= 'size'
)

df_usa_pivot
```

```python
df_usa_pivot.reset_index(inplace=True)

df_usa_pivot
```

```python
df_usa_pivot['month_no'] = pd.to_datetime(df_usa_pivot['job_posted_month'], format='%B')

# out of bounds error can be fixed by including format='%B' parameter; this give us 1900-month-01 format, we can now extract month_no with dt.month

df_usa_pivot['month_no'] = pd.to_datetime(df_usa_pivot['job_posted_month'], format='%B').dt.month

df_usa_pivot
```

```python
df_usa_pivot.set_index('month_no', inplace=True)

df_usa_pivot
```

```python
df_usa_pivot.sort_index(inplace=True)

df_usa_pivot
```

```python
df_usa_pivot.set_index('job_posted_month', inplace=True)

df_usa_pivot
```

```python
df_usa_pivot.plot(kind='line')

# this graph is too congested; let's consider top 3 countries
```

```python
top_3 = df_usa.value_counts('job_title_short').head(3)
top_3 = top_3.index.to_list() # tolist() is used to convert an object into a list
top_3
```

```python
df_usa_pivot[top_3].plot(kind='line')
plt.title('Monthly Job Postings')
plt.xlabel('Month')
plt.ylabel('No. of Job Postings')
plt.legend()
plt.show()
```

```python
# compiled code

import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

df_usa['job_posted_month'] = df_usa['job_posted_date'].dt.strftime('%B') 
df_usa = df[df['job_country'] == 'United States'].copy()

df_usa_pivot = df_usa.pivot_table(
    values= 'salary_year_avg',
    index= 'job_posted_month',
    columns= 'job_title_short',
    aggfunc= 'size'
)

df_usa_pivot.reset_index(inplace=True)
df_usa_pivot['month_no'] = pd.to_datetime(df_usa_pivot['job_posted_month'], format='%B').dt.month
df_usa_pivot.set_index('month_no', inplace=True)
df_usa_pivot.sort_index(inplace=True)
df_usa_pivot.set_index('job_posted_month', inplace=True)
df_usa_pivot.plot(kind='line')

top_3 = df_usa.value_counts('job_title_short').head(3)
top_3 = top_3.index.to_list()

df_usa_pivot[top_3].plot(kind='line')
plt.title('Monthly Job Postings')
plt.xlabel('Month')
plt.ylabel('No. of Job Postings')
plt.legend()
plt.show()
```

```python
# first df is from 6_Pandas_Exercise_Job_Demand

import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

df_usa = df[df['job_country'] == 'United States'].copy()
df_usa['job_posted_month'] = df_usa['job_posted_date'].dt.strftime('%B') 

df_usa_pivot = df_usa.pivot_table(
    values= 'salary_year_avg',
    index= 'job_posted_month',
    columns= 'job_title_short',
    aggfunc= 'size'
)

df_usa_pivot.reset_index(inplace=True)
df_usa_pivot['month_no'] = pd.to_datetime(df_usa_pivot['job_posted_month'], format='%B').dt.month
df_usa_pivot.set_index('month_no', inplace=True)
df_usa_pivot.sort_index(inplace=True)
df_usa_pivot.set_index('job_posted_month', inplace=True)
```

```python
df_usa_pivot
```

```python
# Second df is from "https://lukeb.co/software_csv"
df_software = pd.read_csv("https://lukeb.co/software_csv")

# to get rid of the index, we can code this as:
df_software = pd.read_csv("https://lukeb.co/software_csv", index_col='job_posted_month')

df_software
```

```python
# now let's merge the two dfs

df_usa_merged = df_usa_pivot.merge(df_software, on='job_posted_month')
df_usa_merged
```

```python
df_usa_merged.sum()

# sum() adds up the (size) values corresponding to the index
```

```python
# sorting the values:

top_5 = df_usa_merged.sum().sort_values(ascending=False).head(5)
top_5.index.tolist()

# it can be coded as: top_5 = df_usa_merged.sum().sort_values(ascending=False).head(5).index.tolist()
```

```python
# the above code (since it's long) can be written within () and broken down as follows:

top_5 = (
    df_usa_merged.sum().
    sort_values(ascending=False).
    head(5).
    index.
    tolist()
)

top_5
```

```python
df_usa_merged[top_5].plot(kind='line')
plt.title('Monthly Job Postings Trend')
plt.xlabel('Month')
plt.ylabel('No. of Job Postings')
plt.ylim(0, 22500) # to shift the y-axis limit to 20000 to show the legend properly
plt.legend()
plt.show()
```

## Functions used:
- dictionary comprehension ({month: month for month in months})
- pd.concat('df1', 'df2', ignore_index = True)

```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df

# Create a new column for storing 'Jan-Feb-etc'1_Pandas_Accessing_Data.ipynb

df['job_posted_month'] = df['job_posted_date'].dt.strftime('%b')
# '%b' is the format for 3 letter month values
df
```

```python
# now we create a dictionary comprehension; first the dict:

# dict_months = {
#    'Jan': df_jan,
#    'Feb:' df_feb,
#    'Mar': df_mar,
#    'Apr': df_apr
#    'May': df_may
#    'Jun': df_jun
#    'Jul': df_jul
#    'Aug': df_aug
#    'Sept': df_sep
#    'Oct': df_oct
#    'Nov': df_nov
#    'Dec': df_dec
#}

# create an object to store months

months = df['job_posted_month'].unique()
months


```

```python
{month: month for month in months} # dict comprehension format {key: value for variable in iterable}
```

```python
# we now replace the variable with a filter

{month: df[df['job_posted_month']== month] for month in months}

# result is a mess but it potrays the idea
```

```python
# assign the dict comprehension to a variable 

dict_months = {month: df[df['job_posted_month'] == month] for month in months}

dict_months['Jan']
```

## Boss needs 2023 Q1 data

```python
# we have df_jan, df_feb and df_mar; let's concat them for Q1 data

df_q1 = pd.concat([dict_months['Jan'], dict_months['Feb'], dict_months['Mar']], ignore_index = True)
df_q1
```

```python
# to verify if concat has worked, let's take the value_counts() and visualize it:

df_q1.value_counts('job_posted_month').plot(kind='bar')
plt.title('Job Postings in 2023 - Q1')
plt.xlabel('Month')
plt.ylabel('No. of Job Postings')
plt.xticks(rotation=45)
# plt.text() is used to add data labels
plt.show()
```
